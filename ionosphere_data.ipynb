```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import LabelEncoder
```

# Load Data

```python
data = pd.read_csv("ionosphere_data.csv",delimiter=",")
```

# Split 

```python
train_data = data.values[1:211,0:17].astype(float)
train_label = data.values[1:211,34]
test_data = data.values[211:,0:17].astype(float)
test_label= data.values[211:,34]
```

```python
print(test_data.shape,test_label.shape)
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(train_label)
train_label = encoder.transform(train_label)

encoder = LabelEncoder()
encoder.fit(test_label)
test_label = encoder.transform(test_label)
```

# Model 

```python
model = Sequential()
model.add(Dense(34, input_dim=17 , activation= 'relu' ))
model.add(Dense(1,  activation= 'sigmoid' ))
```

```python
model.fit(train_data, train_label, epochs=100, batch_size=8)
```
